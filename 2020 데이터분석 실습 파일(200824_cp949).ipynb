```python
# 2018년과 2020년 여름 비교하기
```

```python
import csv
f = open('seoul_temp.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
summer2018 = []
summer2020 = []

for row in data :
    year = row[0].split('-')[0]
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '07' or month == '08':
            if year == '2018' : 
                summer2018.append(float(row[-1]))
            if year == '2020' : 
                summer2020.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.figure(dpi = 300)
plt.boxplot([summer2018,summer2020])
plt.show()
```

## 3. 인구데이터(성별 구분 O)

```python
import csv
f = open('age_gender.csv')
data = csv.reader(f)

m = []
f = []
for row in data :
    if '신도림' in row[0] :
        for i in row[3:104] :
            m.append(int(i))
        for i in row[106:] :
            f.append(int(i))
            
print(len(m))
print(len(f))
```

```python
#5. 데이터 시각화하기 : 수평막대그래프(barh) - x축, y축 데이터 모두 명시해야 함!
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)            
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv
f = open('age_gender.csv')
data = csv.reader(f)

m = []
f = []
for row in data :
    if '신도림' in row[0] :
        for i in row[3:104] :
            m.append(-int(i))
        for i in row[106:] :
            f.append(int(i))
            
print(len(m))
print(len(f))
```
